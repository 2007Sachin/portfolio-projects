```python
# First let's import the packages we will use in this project
# You can do this all now or as you need them
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

#Now we need to read the data
df=pd.read_csv(r"C:\Users\sachi\Downloads\movies.csv\movies.csv")
```

```python
df.head()
```

```python
#let's see if there is any missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#Data types for our columns
df.dtypes
```

```python
#filling the NaN Values
df['budget'] = df['budget'].fillna(df['budget'].mean())
df['gross'] = df['gross'].fillna(df['gross'].mean())

```

```python
#change the data type of columns
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
#create correct year column
df['year_correct']=df['released'].astype(str).str[0:4]
df
```

```python
df=df.sort_values(by=['gross'],inplace=False,ascending=False)
```

```python
pd.set_option('display.max_rows',None)
```

```python
#drop any duplicates
df.drop_duplicates()
df.head()
```

```python
#scatter plot with buget vs gross


plt.scatter(x=df['budget'],y=df['gross'])

plt.title('Budget vs Gross Earning')

plt.xlabel('Gross Earning')

plt.ylabel('Budget for Films')

plt.show()
```

```python
correlation_matrix=df.corr(method="spearman")

sns.heatmap(correlation_matrix,annot=True)

plt.title(' correlation matrix for numeric Features')

plt.xlabel('movie Features')

plt.ylabel('movie Features')
```

```python
df_numerized = df.copy()

for col_name in df_numerized.columns:
    if df_numerized[col_name].dtype == 'object':
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
                      
                             
```

```python
correlation_matrix=df_numerized.corr(method="spearman")

sns.heatmap(correlation_matrix,annot=True)

plt.title(' correlation matrix for numeric Features')

plt.xlabel('movie Features')

plt.ylabel('movie Features')
```

```python
df_numerized.corr()
```

```python
correlation_mat=df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs=corr_pairs.sort_values()
sorted_pairs
```
